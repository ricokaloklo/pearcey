```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
from pearcey import pearcey
```

```python
@np.vectorize
def Pe(x, y):
    return pearcey(x, y, algo="numerical")
```

```python
N = 50
xs = np.linspace(2, -10, N)
ys = np.linspace(-10, 10, N)
xx, yy = np.meshgrid(xs, ys, sparse=True)
Pes = Pe(xx, yy)
```

```python
plt.imshow(np.abs(Pes).T, cmap='gray', extent=[ys[0], ys[-1], xs[-1], xs[0]])
plt.xlabel(r"$y$")
plt.ylabel(r"$x$")
plt.title(r"$\mathrm{Pe}(x, y)$")
```
