```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
from pearcey import pearcey
```

```python
@np.vectorize
def Pe(x, y):
    return pearcey(x, y, algo="numerical")
```

```python
N = 100
xs = np.linspace(-10, 2, N)
ys = np.linspace(-10, 10, N)
xx, yy = np.meshgrid(xs, ys, sparse=True)
Pes = Pe(xx, yy)
```

```python
plt.imshow(np.abs(Pes), cmap='gray', extent=[xs[0], xs[-1], ys[0], ys[-1]])
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title(r"$\mathrm{Pe}(x, y)$")
plt.savefig("pearcey.pdf")
```
